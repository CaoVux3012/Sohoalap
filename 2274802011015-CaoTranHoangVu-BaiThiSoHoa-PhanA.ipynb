```python
#Cau 4
import pandas as pd

# T√≠nh correlation cho t·ª´ng Location
correlation_results = df.groupby('Location')[['Hours_Worked', 'Performance']].corr().iloc[0::2, -1]
print("Correlation gi·ªØa Hours_Worked v√† Performance theo Location:")
print(correlation_results)

# K·∫øt qu·∫£ gi·∫£ ƒë·ªãnh (v√¨ kh√¥ng c√≥ d·ªØ li·ªáu th·ª±c t·∫ø):
# Site_A: 0.25
# Site_B: 0.45
# Site_C: 0.65
# Th√™m c·ªôt Work_Allocation d·ª±a tr√™n Location
def allocate_work(location):
    if location == 'Site_A':
        return 'Flexible'
    elif location == 'Site_B':
        return 'Balanced'
    else:
        return 'Time-Intensive'

df['Work_Allocation'] = df['Location'].apply(allocate_work)
print("\nChi·∫øn l∆∞·ª£c ph√¢n b·ªï c√¥ng vi·ªác:")
print(df[['Employee_ID', 'Location', 'Work_Allocation']])
```

```python
#Cau 7
# ƒêi·ªÅn gi√° tr·ªã thi·∫øu Gender b·∫±ng mode
gender_mode = df['Gender'].mode()[0]
df['Gender'] = df['Gender'].fillna(gender_mode)

# T√≠nh t·ª∑ l·ªá nh√¢n vi√™n n·ªØ theo Department
female_ratio = df[df['Gender'] == 'Female'].groupby('Department').size() / df.groupby('Department').size() * 100
print("T·ª∑ l·ªá nh√¢n vi√™n n·ªØ (%):")
print(female_ratio.round(2))
```

```python
#Cau 11
# Lo·∫°i b·ªè c√°c h√†ng c√≥ gi√° tr·ªã thi·∫øu trong Hours_Worked ho·∫∑c Training_Hours ƒë·ªÉ x√¢y d·ª±ng m√¥ h√¨nh h·ªìi quy
df_clean = df.dropna(subset=['Hours_Worked', 'Training_Hours', 'Performance'])

# Chu·∫©n b·ªã d·ªØ li·ªáu cho h·ªìi quy tuy·∫øn t√≠nh
X = df_clean[['Hours_Worked', 'Training_Hours']].values
y = df_clean['Performance'].values

# T√≠nh to√°n h·ªìi quy tuy·∫øn t√≠nh th·ªß c√¥ng (theo c√¥ng th·ª©c: beta = (X^T X)^(-1) X^T y)
X_with_bias = np.c_[np.ones(X.shape[0]), X]  # Th√™m c·ªôt bias (h·ªá s·ªë ch·∫∑n)
beta = np.linalg.inv(X_with_bias.T @ X_with_bias) @ X_with_bias.T @ y

# D·ª± ƒëo√°n gi√° tr·ªã Performance cho c√°c h√†ng c√≥ gi√° tr·ªã thi·∫øu
df_with_missing = df[df['Performance'].isnull()].copy()
X_missing = df_with_missing[['Hours_Worked', 'Training_Hours']].values
X_missing_with_bias = np.c_[np.ones(X_missing.shape[0]), X_missing]
predicted_performance = X_missing_with_bias @ beta

# ƒêi·ªÅn gi√° tr·ªã d·ª± ƒëo√°n v√†o c·ªôt Performance
df.loc[df['Performance'].isnull(), 'Performance'] = predicted_performance
print("D·ªØ li·ªáu sau khi ƒëi·ªÅn gi√° tr·ªã thi·∫øu b·∫±ng h·ªìi quy tuy·∫øn t√≠nh:")
print(df[['Employee_ID', 'Performance']])

# So s√°nh v·ªõi ph∆∞∆°ng ph√°p trung b√¨nh (t√≠nh l·∫°i trung b√¨nh Performance theo Department t·ª´ d·ªØ li·ªáu ban ƒë·∫ßu)
df_mean = pd.DataFrame(data)
mean_performance_by_dept = df_mean.groupby('Department')['Performance'].mean()
df_mean['Performance'] = df_mean.apply(
    lambda row: mean_performance_by_dept[row['Department']] if pd.isnull(row['Performance']) else row['Performance'],
    axis=1
)
print("\nD·ªØ li·ªáu sau khi ƒëi·ªÅn gi√° tr·ªã thi·∫øu b·∫±ng ph∆∞∆°ng ph√°p trung b√¨nh:")
print(df_mean[['Employee_ID', 'Performance']])

# So s√°nh trung b√¨nh Performance c·ªßa t·ª´ng Department gi·ªØa hai ph∆∞∆°ng ph√°p
print("\nTrung b√¨nh Performance theo Department (H·ªìi quy tuy·∫øn t√≠nh):")
print(df.groupby('Department')['Performance'].mean())
print("\nTrung b√¨nh Performance theo Department (Ph∆∞∆°ng ph√°p trung b√¨nh):")
print(df_mean.groupby('Department')['Performance'].mean())

# ƒê·ªÅ xu·∫•t c√°ch ti·∫øp c·∫≠n lai
print("\nƒê·ªÅ xu·∫•t c√°ch ti·∫øp c·∫≠n lai:")
print("- C√°ch ti·∫øp c·∫≠n: K·∫øt h·ª£p h·ªìi quy tuy·∫øn t√≠nh v√† trung b√¨nh b·∫±ng c√°ch s·ª≠ d·ª•ng h·ªìi quy tuy·∫øn t√≠nh ƒë·ªÉ d·ª± ƒëo√°n gi√° tr·ªã thi·∫øu, sau ƒë√≥ ƒëi·ªÅu ch·ªânh gi√° tr·ªã d·ª± ƒëo√°n b·∫±ng c√°ch l·∫•y trung b√¨nh gi·ªØa gi√° tr·ªã h·ªìi quy v√† trung b√¨nh c·ªßa Department.")
print("- V√≠ d·ª•: N·∫øu h·ªìi quy d·ª± ƒëo√°n Performance c·ªßa E002 (IT) l√† 82.5, trung b√¨nh c·ªßa IT l√† 81, th√¨ gi√° tr·ªã cu·ªëi c√πng l√† (82.5 + 81) / 2 = 81.75.")
print("- L√Ω do: H·ªìi quy tuy·∫øn t√≠nh t·∫≠n d·ª•ng m·ªëi quan h·ªá gi·ªØa Hours_Worked, Training_Hours v√† Performance, nh∆∞ng c√≥ th·ªÉ b·ªã ·∫£nh h∆∞·ªüng b·ªüi nhi·ªÖu ho·∫∑c d·ªØ li·ªáu kh√¥ng ƒë·ªß l·ªõn. K·∫øt h·ª£p v·ªõi trung b√¨nh c·ªßa Department gi√∫p gi√° tr·ªã d·ª± ƒëo√°n ·ªïn ƒë·ªãnh h∆°n v√† ph·∫£n √°nh ƒë·∫∑c ƒëi·ªÉm chung c·ªßa ph√≤ng ban.")
```

```python
#Cau 15
import pandas as pd
import numpy as np

def find_outlier_performance(df):
    """
    X√°c ƒë·ªãnh nh√¢n vi√™n c√≥ hi·ªáu su·∫•t n·∫±m ngo√†i 2 ƒë·ªô l·ªách chu·∫©n so v·ªõi gi√° tr·ªã trung b√¨nh.

    Args:
        df (pd.DataFrame): DataFrame ch·ª©a d·ªØ li·ªáu nh√¢n vi√™n, bao g·ªìm c·ªôt 'Performance'.

    Returns:
        pd.DataFrame: DataFrame ch·ª©a th√¥ng tin c·ªßa c√°c nh√¢n vi√™n c√≥ hi·ªáu su·∫•t ngo·∫°i l·ªá.
    """
    performance_cleaned = df['Performance'].dropna()

    if performance_cleaned.empty:
        return pd.DataFrame(columns=df.columns)

    mean_performance = performance_cleaned.mean()
    std_performance = performance_cleaned.std()
    lower_bound = mean_performance - 2 * std_performance
    upper_bound = mean_performance + 2 * std_performance

    outliers_df = df[(df['Performance'] < lower_bound) | (df['Performance'] > upper_bound)]
    return outliers_df

# Gi·∫£ s·ª≠ DataFrame 'df' ƒë√£ ƒë∆∞·ª£c t·∫°o nh∆∞ trong c√°c c√¢u h·ªèi tr∆∞·ªõc
outlier_employees = find_outlier_performance(df.copy())
print("\nNh√¢n vi√™n c√≥ Performance ngo√†i 2 ƒë·ªô l·ªách chu·∫©n:\n", outlier_employees)

print("\nƒê·ªÅ xu·∫•t quy tr√¨nh ƒë√°nh gi√° l·∫°i hi·ªáu su·∫•t cho nh√¢n vi√™n c√≥ Performance ngo√†i 2 ƒë·ªô l·ªách chu·∫©n:")
print("""
Quy tr√¨nh ƒë√°nh gi√° l·∫°i hi·ªáu su·∫•t ƒë∆∞·ª£c ƒë·ªÅ xu·∫•t cho nh·ªØng nh√¢n vi√™n c√≥ Performance n·∫±m ngo√†i 2 ƒë·ªô l·ªách chu·∫©n nh·∫±m m·ª•c ƒë√≠ch hi·ªÉu r√µ h∆°n v·ªÅ nguy√™n nh√¢n d·∫´n ƒë·∫øn k·∫øt qu·∫£ n√†y v√† ƒë∆∞a ra c√°c bi·ªán ph√°p h·ªó tr·ª£ ho·∫∑c ƒëi·ªÅu ch·ªânh ph√π h·ª£p. Quy tr√¨nh n√†y ƒë∆∞·ª£c thi·∫øt k·∫ø ƒë·ªÉ gi·∫£m thi·ªÉu s·ª± thi√™n v·ªã trong qu·∫£n l√Ω:

**B∆∞·ªõc 1: Thu th·∫≠p d·ªØ li·ªáu kh√°ch quan:**

* **Xem x√©t d·ªØ li·ªáu hi·ªáu su·∫•t chi ti·∫øt:** Thay v√¨ ch·ªâ d·ª±a v√†o ƒëi·ªÉm s·ªë Performance t·ªïng th·ªÉ, h√£y xem x√©t c√°c ch·ªâ s·ªë hi·ªáu su·∫•t c·ª• th·ªÉ, c√°c m·ª•c ti√™u ƒë√£ ƒë·∫∑t ra v√† m·ª©c ƒë·ªô ho√†n th√†nh c·ªßa t·ª´ng m·ª•c ti√™u.
* **Thu th·∫≠p d·ªØ li·ªáu v·ªÅ kh·ªëi l∆∞·ª£ng c√¥ng vi·ªác v√† ƒë·ªô ph·ª©c t·∫°p:** ƒê√°nh gi√° xem kh·ªëi l∆∞·ª£ng c√¥ng vi·ªác ƒë∆∞·ª£c giao cho nh√¢n vi√™n c√≥ ph√π h·ª£p kh√¥ng v√† ƒë·ªô ph·ª©c t·∫°p c·ªßa c√°c nhi·ªám v·ª• c√≥ th·ªÉ ·∫£nh h∆∞·ªüng ƒë·∫øn hi·ªáu su·∫•t nh∆∞ th·∫ø n√†o.
* **Xem x√©t d·ªØ li·ªáu v·ªÅ th·ªùi gian l√†m vi·ªác v√† c√°c ngu·ªìn l·ª±c ƒë∆∞·ª£c cung c·∫•p:** ƒê·∫£m b·∫£o r·∫±ng nh√¢n vi√™n c√≥ ƒë·ªß th·ªùi gian v√† c√°c ngu·ªìn l·ª±c c·∫ßn thi·∫øt ƒë·ªÉ ho√†n th√†nh c√¥ng vi·ªác.
* **Thu th·∫≠p ph·∫£n h·ªìi 360 ƒë·ªô (n·∫øu c√≥):** L·∫•y √Ω ki·∫øn ph·∫£n h·ªìi t·ª´ ƒë·ªìng nghi·ªáp, c·∫•p tr√™n, v√† th·∫≠m ch√≠ c·∫£ kh√°ch h√†ng (n·∫øu ph√π h·ª£p) ƒë·ªÉ c√≥ c√°i nh√¨n ƒëa chi·ªÅu v·ªÅ hi·ªáu su·∫•t l√†m vi·ªác c·ªßa nh√¢n vi√™n.

**B∆∞·ªõc 2: Ph·ªèng v·∫•n v√† trao ƒë·ªïi tr·ª±c ti·∫øp:**

* **Ti·∫øn h√†nh cu·ªôc h·ªçp ri√™ng v·ªõi nh√¢n vi√™n:** T·∫°o kh√¥ng gian an to√†n v√† c·ªüi m·ªü ƒë·ªÉ nh√¢n vi√™n chia s·∫ª v·ªÅ nh·ªØng y·∫øu t·ªë c√≥ th·ªÉ ƒë√£ ·∫£nh h∆∞·ªüng ƒë·∫øn hi·ªáu su·∫•t c·ªßa h·ªç (v√≠ d·ª•: kh√≥ khƒÉn trong c√¥ng vi·ªác, thi·∫øu ngu·ªìn l·ª±c, v·∫•n ƒë·ªÅ c√° nh√¢n, v.v.).
* **L·∫Øng nghe ch·ªß ƒë·ªông:** T·∫≠p trung l·∫Øng nghe quan ƒëi·ªÉm c·ªßa nh√¢n vi√™n m√† kh√¥ng ph√°n x√©t hay ng·∫Øt l·ªùi.
* **ƒê·∫∑t c√¢u h·ªèi m·ªü:** Khuy·∫øn kh√≠ch nh√¢n vi√™n chia s·∫ª th√¥ng tin chi ti·∫øt h∆°n v·ªÅ tr·∫£i nghi·ªám v√† nh·∫≠n ƒë·ªãnh c·ªßa h·ªç.

**B∆∞·ªõc 3: Ph√¢n t√≠ch v√† ƒë√°nh gi√° ƒëa chi·ªÅu:**

* **So s√°nh d·ªØ li·ªáu kh√°ch quan v√† √Ω ki·∫øn ph·∫£n h·ªìi:** ƒê·ªëi chi·∫øu c√°c d·ªØ li·ªáu thu th·∫≠p ƒë∆∞·ª£c ·ªü B∆∞·ªõc 1 v√† B∆∞·ªõc 2 ƒë·ªÉ c√≥ c√°i nh√¨n to√†n di·ªán v·ªÅ t√¨nh h√¨nh hi·ªáu su·∫•t c·ªßa nh√¢n vi√™n.
* **Xem x√©t c√°c y·∫øu t·ªë b·ªëi c·∫£nh:** C√¢n nh·∫Øc c√°c y·∫øu t·ªë b√™n ngo√†i c√≥ th·ªÉ ƒë√£ t√°c ƒë·ªông ƒë·∫øn hi·ªáu su·∫•t, ch·∫≥ng h·∫°n nh∆∞ thay ƒë·ªïi trong quy tr√¨nh l√†m vi·ªác, t√°i c·∫•u tr√∫c t·ªï ch·ª©c, ho·∫∑c t√¨nh h√¨nh th·ªã tr∆∞·ªùng.
* **ƒê√°nh gi√° d·ª±a tr√™n ti√™u ch√≠ r√µ r√†ng:** ƒê·∫£m b·∫£o r·∫±ng vi·ªác ƒë√°nh gi√° hi·ªáu su·∫•t ƒë∆∞·ª£c th·ª±c hi·ªán d·ª±a tr√™n c√°c ti√™u ch√≠ ƒë√£ ƒë∆∞·ª£c x√°c ƒë·ªãnh r√µ r√†ng v√† ƒë∆∞·ª£c th√¥ng b√°o tr∆∞·ªõc cho nh√¢n vi√™n.

**B∆∞·ªõc 4: ƒê∆∞a ra quy·∫øt ƒë·ªãnh v√† h√†nh ƒë·ªông:**

* **ƒê·ªëi v·ªõi nh√¢n vi√™n c√≥ hi·ªáu su·∫•t cao b·∫•t th∆∞·ªùng:** T√¨m hi·ªÉu nguy√™n nh√¢n c·ªßa s·ª± v∆∞·ª£t tr·ªôi n√†y v√† xem x√©t vi·ªác c√¥ng nh·∫≠n, khen th∆∞·ªüng ho·∫∑c giao th√™m c√°c th·ª≠ th√°ch m·ªõi ƒë·ªÉ ph√°t huy ti·ªÅm nƒÉng c·ªßa h·ªç.
* **ƒê·ªëi v·ªõi nh√¢n vi√™n c√≥ hi·ªáu su·∫•t th·∫•p b·∫•t th∆∞·ªùng:**
    * **X√°c ƒë·ªãnh nguy√™n nh√¢n g·ªëc r·ªÖ:** D·ª±a tr√™n ph√¢n t√≠ch ·ªü B∆∞·ªõc 3 ƒë·ªÉ x√°c ƒë·ªãnh c√°c y·∫øu t·ªë ch√≠nh d·∫´n ƒë·∫øn hi·ªáu su·∫•t th·∫•p.
    * **X√¢y d·ª±ng k·∫ø ho·∫°ch c·∫£i thi·ªán hi·ªáu su·∫•t:** Thi·∫øt l·∫≠p c√°c m·ª•c ti√™u c·ª• th·ªÉ, ƒëo l∆∞·ªùng ƒë∆∞·ª£c, c√≥ th·ªÉ ƒë·∫°t ƒë∆∞·ª£c, li√™n quan v√† c√≥ th·ªùi h·∫°n (SMART). Cung c·∫•p c√°c ngu·ªìn l·ª±c v√† s·ª± h·ªó tr·ª£ c·∫ßn thi·∫øt ƒë·ªÉ nh√¢n vi√™n c·∫£i thi·ªán.
    * **Theo d√µi v√† ƒë√°nh gi√° ƒë·ªãnh k·ª≥:** Th∆∞·ªùng xuy√™n theo d√µi ti·∫øn ƒë·ªô th·ª±c hi·ªán k·∫ø ho·∫°ch c·∫£i thi·ªán v√† cung c·∫•p ph·∫£n h·ªìi k·ªãp th·ªùi cho nh√¢n vi√™n.

**C√°ch quy tr√¨nh n√†y tr√°nh ƒë∆∞·ª£c thi√™n v·ªã trong qu·∫£n l√Ω:**

* **T·∫≠p trung v√†o d·ªØ li·ªáu kh√°ch quan:** Quy tr√¨nh nh·∫•n m·∫°nh vi·ªác thu th·∫≠p v√† xem x√©t c√°c d·ªØ li·ªáu hi·ªáu su·∫•t c·ª• th·ªÉ, kh·ªëi l∆∞·ª£ng c√¥ng vi·ªác, th·ªùi gian l√†m vi·ªác, v√† c√°c ph·∫£n h·ªìi ƒëa chi·ªÅu, thay v√¨ ch·ªâ d·ª±a v√†o ƒë√°nh gi√° ch·ªß quan c·ªßa ng∆∞·ªùi qu·∫£n l√Ω.
* **Trao ƒë·ªïi tr·ª±c ti·∫øp v√† l·∫Øng nghe:** Vi·ªác t·∫°o c∆° h·ªôi cho nh√¢n vi√™n chia s·∫ª quan ƒëi·ªÉm v√† gi·∫£i th√≠ch v·ªÅ hi·ªáu su·∫•t c·ªßa h·ªç gi√∫p ng∆∞·ªùi qu·∫£n l√Ω hi·ªÉu r√µ h∆°n v·ªÅ c√°c y·∫øu t·ªë c√≥ th·ªÉ ƒë√£ ·∫£nh h∆∞·ªüng ƒë·∫øn k·∫øt qu·∫£, tr√°nh ƒë∆∞·ª£c nh·ªØng ƒë√°nh gi√° phi·∫øn di·ªán.
* **Ph√¢n t√≠ch ƒëa chi·ªÅu v√† xem x√©t b·ªëi c·∫£nh:** Vi·ªác xem x√©t nhi·ªÅu y·∫øu t·ªë kh√°c nhau v√† b·ªëi c·∫£nh l√†m vi·ªác gi√∫p gi·∫£m thi·ªÉu nguy c∆° ƒë√°nh gi√° d·ª±a tr√™n nh·ªØng ƒë·ªãnh ki·∫øn c√° nh√¢n ho·∫∑c nh·ªØng y·∫øu t·ªë kh√¥ng li√™n quan ƒë·∫øn nƒÉng l·ª±c th·ª±c t·∫ø c·ªßa nh√¢n vi√™n.
* **ƒê√°nh gi√° d·ª±a tr√™n ti√™u ch√≠ r√µ r√†ng:** Vi·ªác s·ª≠ d·ª•ng c√°c ti√™u ch√≠ ƒë√°nh gi√° ƒë√£ ƒë∆∞·ª£c x√°c ƒë·ªãnh tr∆∞·ªõc v√† minh b·∫°ch gi√∫p ƒë·∫£m b·∫£o t√≠nh c√¥ng b·∫±ng v√† nh·∫•t qu√°n trong qu√° tr√¨nh ƒë√°nh gi√°.
* **Quy tr√¨nh c√≥ t√≠nh h·ªá th·ªëng:** Vi·ªác tu√¢n theo m·ªôt quy tr√¨nh ƒë√°nh gi√° c·ª• th·ªÉ gi√∫p ƒë·∫£m b·∫£o r·∫±ng t·∫•t c·∫£ c√°c nh√¢n vi√™n c√≥ hi·ªáu su·∫•t ngo·∫°i l·ªá ƒë·ªÅu ƒë∆∞·ª£c xem x√©t m·ªôt c√°ch nh·∫•t qu√°n v√† c√¥ng b·∫±ng.
""")
```

```python
#PHAN B
#Cau 16
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# T·∫£i d·ªØ li·ªáu MNIST v√† chu·∫©n h√≥a
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

# L·∫•y 5 ·∫£nh m·∫´u v√† nh√£n
dataiter = iter(trainloader)
images, labels = next(dataiter) # Use next(dataiter) instead of dataiter.next()

# Chuy·ªÉn c√°c gi√° tr·ªã pixel v·ªÅ [0, 1]
images = images / 2 + 0.5  # ƒê·∫£o ng∆∞·ª£c chu·∫©n h√≥a

# V·∫Ω 5 ·∫£nh m·∫´u v√† nh√£n
fig, axes = plt.subplots(1, 5, figsize=(15, 5))
for i in range(5):
    axes[i].imshow(images[i].numpy().squeeze(), cmap="gray")
    axes[i].set_title(f'Label: {labels[i].item()}')
    axes[i].axis('off')

plt.show()

# ƒê·ªÅ xu·∫•t c√°ch tr·ª±c quan h√≥a d·ªØ li·ªáu kh√°c: Bi·ªÉu ƒë·ªì ph√¢n ph·ªëi c√°c gi√° tr·ªã pixel
pixels = images.numpy().reshape(-1, 784)  # Chuy·ªÉn th√†nh vector 784 chi·ªÅu
pixels = pixels.flatten()

# V·∫Ω histogram c·ªßa gi√° tr·ªã pixel
plt.figure(figsize=(7, 5))
plt.hist(pixels, bins=50, color='gray')
plt.title("Distribution of Pixel Values")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

# T·∫£i d·ªØ li·ªáu MNIST
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

# ƒê·ªãnh nghƒ©a ki·∫øn tr√∫c ANN
class ANN_Model(nn.Module):
    def __init__(self):
        super(ANN_Model, self).__init__()
        self.fc1 = nn.Linear(784, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32, 10)

    def forward(self, x):
        x = x.view(-1, 784)  # Flatten 28x28 images to 784
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x

# Kh·ªüi t·∫°o m√¥ h√¨nh
model = ANN_Model()

# Loss function v√† optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Hu·∫•n luy·ªán m√¥ h√¨nh v·ªõi early stopping
epochs = 20
patience = 5
best_loss = float('inf')
patience_counter = 0
train_loss = []

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for images, labels in trainloader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    avg_loss = running_loss / len(trainloader)
    train_loss.append(avg_loss)

    print(f'Epoch {epoch+1}/{epochs}, Loss: {avg_loss}')

    # Early stopping
    if avg_loss < best_loss:
        best_loss = avg_loss
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print("Early stopping triggered")
            break

# V·∫Ω bi·ªÉu ƒë·ªì loss
plt.plot(train_loss)
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()

```

```python
import torch
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms


# Define the test dataset and dataloader
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
testset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=False)

# D·ª± ƒëo√°n tr√™n t·∫≠p ki·ªÉm tra
model.eval()  # ƒê·∫∑t m√¥ h√¨nh ·ªü ch·∫ø ƒë·ªô ƒë√°nh gi√°
correct = 0
total = 0
all_labels = []
all_preds = []

with torch.no_grad():
    for images, labels in testloader:  # testloader l√† DataLoader c·ªßa t·∫≠p ki·ªÉm tra
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        all_labels.extend(labels.numpy())
        all_preds.extend(predicted.numpy())

accuracy = 100 * correct / total
print(f"Accuracy on test set: {accuracy}%")

# T√≠nh confusion matrix
cm = confusion_matrix(all_labels, all_preds)

# V·∫Ω confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
